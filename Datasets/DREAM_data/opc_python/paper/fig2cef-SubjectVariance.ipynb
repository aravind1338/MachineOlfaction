```python
# Preliminaries to work with the data.   
%matplotlib inline
%run __init__.py # Add the current directory to the path.  
from utils import loading, scoring
from gerkin import dream,params
import numpy as np
import matplotlib.pyplot as plt
import pandas
```

## Subject Clustering

```python
# Load the observed perceptual descriptor data for the high concentration (low dilution) of each pair.  
data = loading.load_perceptual_data(['training','leaderboard','testset'])
data = dream.filter_Y_dilutions(data,'high')
descriptors = loading.get_descriptors(format=True)
```

```python
# Compute means and stdevs
y = data.unstack('Descriptor').T
means = y.mean(axis=1).unstack()[descriptors].values
stdevs = y.std(axis=1).unstack()[descriptors].values
y_moments = np.hstack((means,stdevs))
print("%d (# subjects) by %d (2 times the # of descriptors)" % y_moments.shape)
```

```python
# Cluster the data.  
from scipy.cluster.hierarchy import fclusterdata,dendrogram,linkage
fclusterdata(y_moments,1)
Z = linkage(y_moments, 'ward')
```

```python
# Create the dendrogram
plt.figure(figsize=(10, 4))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Subject number')
plt.ylabel('Distance')
d = dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
    distance_sort = 'ascending'
)

# Create a copy of the data that is sorted by subjection position in the dendrogram.
y_dend = y_moments.copy()
for i in range(49):
    y_dend[i,:] = y_moments[int(d['ivl'][i]),:]
```

```python
# Swap the red and cyan branch for visualization since the ordering is arbitrary
divl_ = np.array(d['ivl'])
divl_[5:30] = d['ivl'][24:]
divl_[30:] = d['ivl'][5:24]
y_dend_ = y_dend.copy()
for i in range(49):
    y_dend_[i,:] = y_moments[int(divl_[i]),:]
y_dend = y_dend_
```

```python
# Create a version of the same that is normalized to the mean value for each descriptor.  
y_dend_norm = y_dend / np.mean(y_dend,axis=0,keepdims=True)
```

## Fig. 2C and some auxiliary plots

### Mean vs StDev for each descriptor; each point is one subject (Colors are as in dendrogram)

```python
import matplotlib as mpl
mpl.rcParams.update({'font.size': 12})
fig,axes = plt.subplots(3,7,figsize=(15,8))
colors = ['k']*49
for i in range(0,5):
    subject = int(d['ivl'][i])
    colors[subject] = 'green'
for i in range(5,27):
    subject = int(d['ivl'][i])
    colors[subject] = 'red'
for i in range(27,49):
    subject = int(d['ivl'][i])
    colors[subject] = 'cyan'
for i,ax in enumerate(axes.flat):
    ax.scatter(y_moments[:,i],y_moments[:,21+i],color=colors)
    ax.set_xlim(0,100)
    ax.set_ylim(0,50)
    ax.set_title(descriptors[i].split('/')[0])
plt.tight_layout()
```

```python
# Load predictions from previously run models
from scipy.io import matlab
yg = matlab.loadmat('../../data/sc2_yg.mat') # Load Yuanfang's predictions.  
yg = yg['a2']
resort = [sorted([str(i) for i in range(1,50)]).index(str(s)) for s in range(1,50)]
yg = yg[:,resort]
#y = np.ma.dstack([Y['subject'][i] for i in range(1,50)])

rg = np.load('../../data/rg.npy') # Load Rick's predictions.  

# Take the mean of these two sets of predictions
pred = rg*0.5 + yg*0.5
```

### Prediction quality vs StDev for each descriptor; each point is one subject (Colors are as in dendrogram)

```python
mpl.rcParams.update({'font.size': 12})
rs = np.zeros(21)
fig,axes = plt.subplots(3,7,figsize=(15,9))        
for i,ax in enumerate(axes.flat):
    y = data.loc[descriptors[i]]
    ax.scatter(y.std(axis=0),pred[i,:],color=colors)
    ax.set_xlim(0,50)
    ax.set_ylim(0,1)
    ax.set_xticks([0,10,20,30,40,50])
    rs[i] = np.ma.corrcoef(y.std(axis=0),pred[i,:])[1,0]
    ax.set_title('%s\nR=%.2f' % (descriptors[i].split('/')[0],rs[i]))
plt.tight_layout()
axes.flat[7].set_ylabel('Correlation between data and prediction',size=16,labelpad=20)
axes.flat[17].set_xlabel('Data standard deviation',size=16,labelpad=20);
```

#### Compute two-tailed for $\alpha=0.05$

```python
alpha = 0.05
n_iter = 1000
n_desc = len(descriptors)
n_subjects = 49
rs_shuffle = np.zeros((n_desc,n_iter))
for j in range(n_iter):
    randos = np.argsort(np.random.random(n_subjects))
    for i,descriptor in enumerate(descriptors):
        y = data.loc[descriptor]
        rs_shuffle[i,j] = np.ma.corrcoef(y.std(axis=0),pred[i,randos])[1,0]
threshold = sorted(rs_shuffle.ravel())[int((1.0 - alpha/2)*n_desc*n_iter)]
print("Threshold for p<0.05 is %.3f" % threshold)
```
